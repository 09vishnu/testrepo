## 2. Data Cleaning (5 Marks)

### 2.1 Get the column-wise Null percentage count using 'isnull()' and length of the dataframe (2 Marks)

#### Hint: When we use isnull().sum() we get sum and when we divide it by the length of the dataframe and multiply the numerator with 100 we get the percentage of missing values in each column

```python
# Get the percentage of missing values in each column. Sort the output in descending order
Null_percentage = hotel.isnull().sum()/hotel.shape[0]*100
Null_percentage_descending = Null_percentage.sort_values(ascending=False)
print(Null_percentage_descending)

# isnull().sum() is used to get the column wise Null count of this dataframe. 
# To obtain percentage, the sum value has to be divided by the length of the dataframe and multiply with 100.
```

### 2.2 We can see that the columns 'company' and 'agent' have 94% and 13% of values missing respectively. Since our business goals doesnot require these columns also, since these columns contain huge missing values.So,lets drop them. (2 marks)

```python
 # Use '.drop' and drop the columns
hotel = hotel.drop(['company', 'agent'], axis = 1)
hotel.head(3)

# To remove columns we need to mention axis = 1. Axis 1 represents the column and Axis 0 represents the rows.
# By default the drop function will consider as axis = 0.
```

```python
hotel.shape

# Shape returns total no.of rows and columns in the dataframe. No. of columns has reduced from 33 to 31.
```

### Though country contains 0.40% of missing values it is a important factor in our analysis. Also, dropping the whole column for such low missing values is not the right choice. Lets drop the rows which has column 'country' missing.

```python
hotel = hotel.dropna(subset=['country'])
hotel.head(3)

# By default dropna will drop rows with null value. Therefore to drop null values in the specific column.
# It has to be specified as subset within the dropna()
```

```python
hotel.shape

# Now the no.of rows got reduced from 119390 to 118902.
```

```python
# Removed all the null values in Country Column - Data cleaning is done.
hotel.isnull().sum()

# This gives the sum of all null values in the dataframe. Now the Country column doesn't have null values. 
```

## 3.2 Filtering out top 10 customers to offer them discounts when they visit next time (14 Marks)

### 3.2.1 Filter customers that have not cancelled the booking. (2 marks)

```python
not_cancel = hotel[hotel['is_canceled']==0]
not_cancel

# This is a new dataframe called not_cancel and with not cancelled filtered from the Hotel Dataframe. 
```

From now on we will use this dataframe for further analysis

Let's now calculate the total stay duration for customer together with the revenue generated by that particular customer. Using these metrics, we would be able to identify top customers that are profitable to the hotel. 

### 3.2.2 Create a new column "total_stay_duration" in the data frame not_cancel. (2 Marks)

Total_Stay_Duration is defined as the number of nights a customer stayed in the hotel including both weekends and weekdays

```python
not_cancel['total_stay_duration'] = not_cancel['stays_in_weekend_nights'] + not_cancel['stays_in_week_nights']
not_cancel
    
# New coloumn total stay duration is added to the not_cancel dataframe with new value added to it.
```

### 3.2.3 Create a new column "total_revenue" in the data frame not_cancel. (2 Marks)

Total_Revenue can be calculated by multiplying average rental revenue(adr) with total stay duration(column created in last question)

The average daily rate (adr) measures the average rental revenue earned for an occupied room per day. The operating performance of a hotel or other lodging business can be determined by using the ADR.

```python
not_cancel['Total_revenue'] = not_cancel['adr'] * not_cancel['total_stay_duration']
not_cancel

# Another new column is added to the dataframe called Total_revenue and obtained values multiplied by adr and total stay duration.
```

```python
not_cancel.Total_revenue.sum()
```
